```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Bar plot

```python
df = pd.read_csv('online_food_order.csv')
plt.figure(figsize=(12, 6))
sns.barplot(x='TotalAmount', y='FoodItem', data=df, estimator=np.mean, ci=None, palette='coolwarm')
plt.title('Average Total Amount by Food Item')
plt.show()
```

# Pie chart

```python
average_total_by_food = df.groupby('FoodItem')['TotalAmount'].mean().sort_values(ascending=False)

# Plotting
plt.figure(figsize=(8, 8))
plt.pie(average_total_by_food, labels=average_total_by_food.index, autopct='%1.1f%%', startangle=140)
plt.title('Average Total Amount by Food Item')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

# Box plot

```python
plt.figure(figsize=(12, 6))
sns.boxplot(y='RestaurantName', x='DeliveryTime', data=df)
plt.title('Delivery Time by Restaurant')
plt.show()
```

# Count plot


```python
sns.countplot(x='Rating', data=df, palette='viridis')
plt.title('Ratings Distribution')
plt.show()
```

```python
# Example: K-means clustering for customer segmentation
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler

# Select relevant features for clustering
X = df[['TotalAmount', 'DeliveryTime', 'Rating']]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Determine optimal number of clusters using Elbow Method
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='--')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
feedback_counts = df['Feedback'].value_counts()

labels = feedback_counts.index
sizes = feedback_counts.values
explode = (0.1,) * len(labels)  # explode each slice to highlight them

plt.figure(figsize=(8, 6))
plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Overall Customer Feedback Analysis')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```
